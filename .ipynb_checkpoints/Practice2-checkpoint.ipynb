```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
linear_data=np.arange(1,8)
quadratic_data=linear_data**2
```

```python
x=np.arange(1,8)
```

```python
%matplotlib notebook
```

```python
plt.figure()
s1=plt.subplot(1,2,1)

s1.plot(x,linear_data,'-r',label='linear data')
plt.show()
```

```python
s2=plt.subplot(1,2,2)
s2.plot(x,quadratic_data,'-b',label='quadratic data')
plt.show()
```

```python
s1.set_xlabel('X')
s1.set_ylabel('Linear Data')
s1.set_title('Linear Data')
s1.legend()

s2.set_xlabel('X')
s2.set_ylabel('quadratic Data')
s2.set_title('Quadratic Data')
s2.legend()
```
