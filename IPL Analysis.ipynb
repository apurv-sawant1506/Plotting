```python
df.drop(['id','umpire1','umpire2','umpire3'],axis=1,inplace=True)
```

```python
winning_rows=df[df['toss_winner']==df['winner']].shape[0]
```

```python
total_rows=df.shape[0]
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib notebook
```

```python
plt.figure()
labels=['yes','no']
sizes=[winning_rows,total_rows-winning_rows]
colors=['yellow','blue']
plt.pie(sizes,labels=labels,colors=colors,autopct='%1.1f%%', startangle=90)
plt.axis('equal')
plt.title('Toss Winner is match winner')
plt.show()
```

```python
import seaborn as sns
```

```python
list1=[]
for i in range(len(df)):
    if(df['toss_winner'][i]==df['winner'][i]):
        list1.append('yes')
    else:
        list1.append('no')

df['toss_winner_is_winner']=list1
```

```python
df.head()
```

### The Below visual shows winning count after winning toss and after losing toss per team

```python
plt.figure(figsize=(8,6))
sns.countplot(x='toss_winner', hue='toss_winner_is_winner', data=df)
plt.xticks(rotation='vertical')
plt.subplots_adjust(bottom=0.45)
plt.show()
```
